```python
class Prevodnik:

  """
  Rodičovská třída pro převod čísel mezi soustavami od dvojkové do šestnáctkové.
  """

  def __init__(self, cislo):
    self._cislo = cislo

  def vyber(self):

      """
      Vybírá konkrétní převod mezi soustavami.
      """
      vyber_prevodu = None
      while vyber_prevodu not in(1, 2, 3, ""):
        print("""
        Vyber:
        1) pro převod z desítkové soustavy,
        2) pro převod do desítkové soustavy,
        3) pro převod mezi různými soustavami,
           pro konec dejte enter.
        """)
        vyber_prevodu=input("Vyber převod: ")
        if vyber_prevodu == "1":
            print("Převod z desítkové soustavy.")
            soustava = self.prevod_z_desitkove()
            print(f"Převod čísla {self._cislo} ze soustavy 10 do soustavy {soustava} je", end=" ")
            vysledek = Prevodnik_z_desitkove.prevod_z_desitkove(self, self._cislo, soustava)
            print(f"{vysledek}.")
            break
        if vyber_prevodu == "2":
            print("Převod do desítkové soustavy.")
            soustava = self.prevod_do_desitkove()
            print(f"Převod čísla {self._cislo} ze soustavy {soustava} do soustavy 10 je", end=" ")
            vysledek = Prevodnik_do_desitkove.prevod_do_desitkove(self, soustava)
            print(f"{vysledek}.")
            break
        if vyber_prevodu == "3":
            print("Převod mezi různými soustavami.")
            soustava = self.prevod_ostatni(cislo)
            ze_soustavy = soustava[0]
            do_soustavy = soustava[1]
            print(f"Převod čísla {self._cislo} ze soustavy {ze_soustavy} do soustavy {do_soustavy} je", end=" ")
            vysledek = Prevodnik_ostatni.prevod_ostatni(self, ze_soustavy, do_soustavy)
            print(f"{vysledek}.")
            break
        if vyber_prevodu == "":
            print("konec")
            break

  def prevod_z_desitkove(self):

    """
    Vrací soustavu pro převod z desítkové.
    """
    soustava = int(input("Vyber soustavu, do které převádíš (2 - 16): "))
    while soustava not in list(range(2, 17)):
      print(f"Vybrali jste chybnou soustavu {soustava}.")
      soustava = int(input("Vyber soustavu 2 - 16: "))
    print(f"Vybrali jste převod do soustavy {soustava}.")

    return soustava

  def prevod_do_desitkove(self):

    """
    Vrací soustavu pro převod do desítkové.
    """
    soustava = int(input("Vyber soustavu, do které převádíš (2 - 16): "))
    while soustava not in list(range(2, 17)):
      print(f"Vybrali jste chybnou soustavu {soustava}.")
      soustava = int(input("Vyber soustavu 2 - 16: "))
    print(f"Vybrali jste převod do soustavy {soustava}.")

    return soustava

  def prevod_ostatni(self, cislo):

    """
    Vrací dvě soustavy:
    1) z které převádí
    2) do které převádí.
    """

    if cislo[0].isalpha():
      ze_soustavy = int(input("Vyber soustavu, z které převádíš (10 - 16): "))
      while ze_soustavy not in list(range(10, 17)):
        print(f"Vybrali jste chybnou soustavu {ze_soustavy}.")
        ze_soustavy = int(input("Vyber soustavu 10 - 16: "))
    else:
      ze_soustavy = int(input("Vyber soustavu, z které převádíš (2 - 16): "))
      while ze_soustavy not in list(range(2, 17)):
        print(f"Vybrali jste chybnou soustavu {ze_soustavy}.")
        ze_soustavy = int(input("Vyber soustavu 2 - 16: "))
    print(f"Vybrali jste převod ze soustavy {ze_soustavy}.")

    do_soustavy = int(input("Vyber soustavu, do které převádíš (2 - 16): "))
    while do_soustavy not in list(range(2, 17)):
      print(f"Vybrali jste chybnou soustavu {do_soustavy}.")
      do_soustavy = int(input("Vyber soustavu 2 - 16: "))
    print(f"Vybrali jste převod do soustavy {do_soustavy}.")

    return ze_soustavy, do_soustavy

class Prevodnik_z_desitkove(Prevodnik):

  """
  Třída převádí čísla z desítkové soustavy.
  """
  def __init__(self, cislo):
    self._cislo = cislo

  def prevod_z_desitkove(self, cislo, soustava):

    """
    Převádí čísla z desítkové soustavy.
    """
    vysledek = []
    podil = None
    cislo = int(cislo)
    while podil != 0:
      podil = cislo // soustava
      zbytek = cislo - podil * soustava
      if zbytek < 10:
        vysledek.append(zbytek)
      else:
        pismena_ascii = []
        cisla_ascii = list(range(65, 91))
        for i in cisla_ascii:
          pismena_ascii.append(chr(i))
        ind=(zbytek - 9) - 1
        vysledek.append(pismena_ascii[ind])
      cislo = podil
    vystup = ""
    for i in range(len(vysledek)-1, -1, -1):
      vystup += str(vysledek[i])

    return vystup

class Prevodnik_do_desitkove(Prevodnik):

  """
  Třída převádí čísla do desítkové soustavy.
  """
  def __init__(self, cislo):
    self._cislo = cislo

  def prevod_do_desitkove(self,soustava):

    """
    Převádí čísla do desítkové soustavy.
    """
    cislo_revers = self._cislo[::-1]
    pismena_ascii = []
    cisla_ascii = list(range(65, 91))
    for j in cisla_ascii:
      pismena_ascii.append(chr(j))
    vysledek = 0

    for i in range(len(cislo_revers)-1, -1, -1):
      if cislo_revers[i].isdigit():
        vysledek += int(cislo_revers[i]) * soustava ** i
      else:
        ind = pismena_ascii.index(cislo_revers[i]) + 1
        cislo = 9 + ind
        vysledek += cislo * soustava ** i

    return vysledek

class Prevodnik_ostatni(Prevodnik_do_desitkove):

  """
  Třída převádí čísla mezi různými soustavami.
  """
  def __init__(self, cislo):
    self.cislo = cislo

  def prevod_ostatni(self, ze_soustavy, do_soustavy):

    """
    Převádí čísla mezi různými soustavami.
    """
    do_desitkove = Prevodnik_do_desitkove.prevod_do_desitkove(self, ze_soustavy)
    z_desitkove = Prevodnik_z_desitkove.prevod_z_desitkove(self, do_desitkove, do_soustavy)
    return z_desitkove

cislo = input("Zadej číslo k převodu: ").upper()
prevodnik = Prevodnik(cislo)
prevodnik.vyber()
```
